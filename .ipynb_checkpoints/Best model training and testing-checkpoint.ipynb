## Best SVM and MLP model testing

```python
# importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import time as time
```

```python
# reading datasets

X_enc = pd.read_csv("X_enc.csv")
X_enc = X_enc.drop("Unnamed: 0", axis = 1)
y = pd.read_csv("y.csv")
y = y.drop("Unnamed: 0", axis = 1)
```

```python
# Using a train test split with 75%/25% train/test ratio

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_enc, y, test_size=0.25, random_state = 1)
```

```python
# Scaling data after train test split

from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()

X_trainS = min_max_scaler.fit_transform(X_train)
X_testS = min_max_scaler.fit_transform(X_test)
```

```python
#Implementing undersampling with sampling strategy 0.5
#Borderline SMOTE implement to balance training set

from imblearn.over_sampling import BorderlineSMOTE
from imblearn.under_sampling import RandomUnderSampler 
undersample = RandomUnderSampler(sampling_strategy=0.99)
oversample = BorderlineSMOTE()

X_trainres, y_trainres = undersample.fit_resample(X_trainS, y_train)
X_trainres, y_trainres = oversample.fit_resample(X_trainres, y_trainres)
```

```python
X_trainres = pd.DataFrame(X_trainres, columns=X_enc.columns)
X_testS = pd.DataFrame(X_testS, columns=X_enc.columns)
```

### Training optimal MLP

```python
#Importing torch libraries
import torch
import torch.nn as nn
```

```python
#Converting data to tensors

X_train_tensor = torch.tensor(X_trainres.astype('float32').values)
y_train_tensor = torch.tensor(y_trainres.astype('float32').values)


X_test_tensor = torch.tensor(X_testS.astype('float32').values)
y_test_tensor = torch.tensor(y_test.values)
y_test_tensor = y_test_tensor.reshape(1,-1)[0]
```

```python
import torch.nn.functional as F

# Training model on optimal hyperparameters 

class FinalNet(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, dropout = 0.5):
        
        super(FinalNet, self).__init__()
        
        self.input = nn.Linear(input_dim, hidden_dim)
        self.hidden1 = nn.Linear(hidden_dim, int(hidden_dim/2))
        self.output = nn.Linear(int(hidden_dim/2), output_dim)
        
        self.dropout = nn.Dropout(dropout)
    
    def forward(self, X):
        
        input = self.input(X)
        input = F.relu(input)
        
        hidden1 = self.hidden1(input)
        hidden1 = F.relu(hidden1)
        
        output = self.output(hidden1)
        output = self.dropout(output)
        output = F.softmax(output, dim = -1)
        
        return output
```

```python
# Building network with skorch on optimal found hyperparameters

from skorch import NeuralNetClassifier
from skorch.callbacks import EarlyStopping

# Defining the number of neurons for each layer
inc_dim = X_trainres.shape[1] # Input layer
output_dim = 2 # Output layer, 2 because of 2 classes

FNet = NeuralNetClassifier(
    FinalNet(inc_dim, 400, output_dim),
    max_epochs= 1000,
    callbacks=[EarlyStopping()],
    optimizer = torch.optim.Adam,
    criterion = nn.NLLLoss,
    lr = 0.00001
)
```

```python
# training MLP and logging time taken

mlpt0 = time.time()
FNet.fit(X_train_tensor, y_train_tensor.reshape(1,-1)[0].long())
mlpt1 = time.time()
```

```python
print("MLP training time taken: ", mlpt1-mlpt0,"seconds")
```

```python
## Saving model with best hyperparameters
import pickle

filename_mlp = 'Best_MLP.sav'
pickle.dump(FNet, open(filename_mlp, 'wb'))
```

### Training optimal SVM

```python
from sklearn.svm import SVC

SVM = SVC(C=100, gamma = 1, kernel = 'rbf',probability = True)
```

```python
# training svm and logging time taken

svmt0 = time.time()
SVM.fit(X_trainres, y_trainres.values.ravel())
svmt1 = time.time()
```

```python
print("SVM training time taken: ", svmt1-svmt0, "seconds")
```

```python
# Saving best SVM 
import pickle

filename_svm = 'Best_SVM.sav'
pickle.dump(SVM, open(filename_svm, 'wb'))
```

## Testing best models

### MLP 

```python
# Loading best MLP
best_MLP = pickle.load(open(filename_mlp, 'rb'))
```

```python
# Evaluation metrics

from sklearn.metrics import confusion_matrix

y_pred_mlp = best_MLP.predict(X_test_tensor)
cm_mlp = confusion_matrix(y_test_tensor, y_pred_mlp)

print(cm_mlp)

print("Accuracy = ", (cm_mlp[0,0]+cm_mlp[1,1])/cm_mlp.sum())

print("F1-Score = ", cm_mlp[0][0]/(cm_mlp[0][0] + (1/2)*(cm_mlp[1][0]+cm_mlp[1][1])))

print("Sensitivity = ", cm_mlp[0][0]/(cm_mlp[0][1]+cm_mlp[0][0])) 

print("Specificity = ", cm_mlp[1][1]/(cm_mlp[1][0]+cm_mlp[1][1]))
```

```python
# confusion matrix heatmap

import seaborn as sns
sns.heatmap(cm_mlp, annot=True, cmap='Blues', fmt='g')
plt.title("MLP")
```

```python
y_pred_prob_mlp = FNet.predict_proba(X_test_tensor)[:,1] #Class probabilities
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
auc_mlp = roc_auc_score(y_test, y_pred_mlp) # AUC calculation
```

```python
fpr_mlp, tpr_mlp, _ = roc_curve(y_test, y_pred_prob_mlp) # ROC calculation
```

```python
# plotting roc curve
plt.plot(fpr_mlp, tpr_mlp, color='black', label='MLP, AUC = {}'.format(round(auc_mlp,4)))
plt.plot([0,1],[0,1], linestyle='dashed')
plt.xlabel("False-positive rate")
plt.ylabel("True-positive rate")

plt.legend()
plt.title("ROC curve")
```
