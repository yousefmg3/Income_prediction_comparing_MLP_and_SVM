```python
# Plotting histograms of numeric data to see distribution 

col = inc.select_dtypes(include=np.number).columns.tolist()[:-1]
fig1, axs = plt.subplots(2,3,figsize=(15,10))
axs = axs.ravel()
for i in range(len(col)):
    axs[i].hist(inc[col[i]],bins = 20, edgecolor='black', facecolor='#eab9b4')
    axs[i].set_xlabel(col[i])
    axs[i].set_ylabel("Frequency")
   # axs[i].set_title("S = {}, K = {}".format(round(skew(country_data[col[i]]),3),round(kurtosis(country_data[col[i]]),3)))
                                                       
plt.savefig("histograms.png")
```

Data is heavily skewed, data should be scaled 

```python
# Using z-score, datapoints are removed if they are <-3 or >3

col = inc.select_dtypes(include=np.number).columns.tolist()[:-1]
z_scores = stats.zscore(inc[col])
#calculate z-scores of `df`
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
new_inc = inc[filtered_entries].reset_index().drop("index",axis=1)
```

Outliers removed since there are not many, 40727 before removing outliers, 37250 after, not a significant amount of outliers so safe to remove

```python
# Checking if data is still heavily skewed after removing outliers and null values
new_inc.groupby("income_>50K").agg(count=("income_>50K","count"))
```

```python
# Data contains a lot of categorical columns, one hot encoding used
# Data split into features and target variable (X and y) 
X = new_inc.loc[:, inc.columns != "income_>50K"]
X_enc = pd.get_dummies(data=X, columns=["workclass","education","marital-status","occupation","relationship","race","gender","native-country"],prefix='',prefix_sep='')
y = new_inc[["income_>50K"]]
```

### SMOTE implementation and saving training and testing data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

plt.style.use("seaborn")
```

```python
# Using a train test split with 75%/25% train/test ratio
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_enc, y, test_size=0.25, random_state = 1)
```

```python
# Scaling data after train test split
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()

X_trainS = min_max_scaler.fit_transform(X_train)
X_testS = min_max_scaler.fit_transform(X_test)
```

```python
#Implementing undersampling with sampling strategy 0.5
#Borderline SMOTE implement to balance training set

from imblearn.over_sampling import BorderlineSMOTE
from imblearn.under_sampling import RandomUnderSampler 
undersample = RandomUnderSampler(sampling_strategy=0.5)
oversample = BorderlineSMOTE()

X_trainres, y_trainres = undersample.fit_resample(X_trainS, y_train)
X_trainres, y_trainres = oversample.fit_resample(X_trainres, y_trainres)
```

```python
X_trainres = pd.DataFrame(X_trainres, columns=X_enc.columns)
X_testS = pd.DataFrame(X_testS, columns=X_enc.columns)
```

```python
y_trainres.groupby("income_>50K").agg(count=("income_>50K","count")) # Making sure training classes are balanced
```

### 1 hidden layer

```python
#Importing torch libraries
import torch
import torch.nn as nn
```

```python
# Defining the number of neurons for each layer
inc_dim = X_trainres.shape[1] # Input layer
hidden_dim = inc_dim*2 # Hidden layer
output_dim = 2 # Output layer, 2 because of 2 classes
```

```python
#Converting data to tensors

X_train_tensor = torch.tensor(X_trainres.astype('float32').values)
y_train_tensor = torch.tensor(y_trainres.astype('float32').values)


X_test_tensor = torch.tensor(X_testS.astype('float32').values)
y_test_tensor = torch.tensor(y_test.values)
y_test_tensor = y_test_tensor.reshape(1,-1)[0]
```

```python
import torch.nn.functional as F

#Initial testing using 1 hidden layer and sigmoid activation function

class NN_sigmoid(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, dropout = 0.5):
        
        super(NN_sigmoid, self).__init__()
        
        self.input = nn.Linear(input_dim, hidden_dim)
        self.hidden = nn.Linear(hidden_dim, int(hidden_dim/2))
        self.output = nn.Linear(int(hidden_dim/2), output_dim)
        
        self.dropout = nn.Dropout(dropout)
        self.sigmoid = torch.nn.Sigmoid()
    
    def forward(self, X):
        
        input = self.input(X)
        input = self.sigmoid(input)
        
        hidden = self.hidden(input)
        hidden = self.sigmoid(hidden)
        
        output = self.output(hidden)
        output = self.dropout(output)
        output = F.softmax(output, dim = -1)
        
        return output
```

```python
from skorch import NeuralNetClassifier
from skorch.callbacks import EarlyStopping
```

```python
#Using skorch to train the network

net_sigmoid = NeuralNetClassifier(
    NN_sigmoid(inc_dim, hidden_dim, output_dim),
    max_epochs=100,
    callbacks=[EarlyStopping()]
)
```

```python
from sklearn.model_selection import GridSearchCV

# Carrying out gridsearch to find with learning rate and momentum on sigmoid activation function that will be compared
# with relu activation function

param_grid = {'lr': [0.1,0.01, 0.001, 0.0001], 'optimizer__momentum': [0.9,0.7,0.5,0.3,0.1]}
grid_sigmoid = GridSearchCV(net_sigmoid, param_grid, scoring = , refit=True, n_jobs = -1)
```

```python
grid_sigmoid.fit(X_train_tensor, y_train_tensor.reshape(1,-1)[0].long())
```

### Grid search with optimizer and loss criterion

```python
import torch.nn.functional as F

class NN(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, dropout = 0.5):
        
        super(NN, self).__init__()
        
        self.input = nn.Linear(input_dim, hidden_dim)
        self.hidden = nn.Linear(hidden_dim, int(hidden_dim/2))
        self.output = nn.Linear(int(hidden_dim/2), output_dim)
        
        self.dropout = nn.Dropout(dropout)
    
    def forward(self, X):
        
        input = self.input(X)
        input = F.relu(input)
        
        hidden = self.hidden(input)
        hidden = F.relu(hidden)
        
        output = self.output(hidden)
        output = self.dropout(output)
        output = F.softmax(output, dim = -1)
        
        return output
```

```python
net_gs = NeuralNetClassifier(
    NN(inc_dim, hidden_dim, output_dim),
    max_epochs=300,
    callbacks=[EarlyStopping()],
)
```

```python
from sklearn.model_selection import GridSearchCV

# Grid search for learning rate, optimizer and loss criterion

param_grid = {'lr': [0.1,0.01, 0.001, 0.0001, 0.00001], 'optimizer': [torch.optim.SGD, torch.optim.Adam], 'criterion': [nn.CrossEntropyLoss, nn.NLLLoss]}
gridNN = GridSearchCV(net_gs, param_grid, refit=True, n_jobs = -1)
```

```python
gridNN.fit(X_train_tensor, y_train_tensor.reshape(1,-1)[0].long())
```

```python
gridNN.best_score_
```

```python
gridNN.best_params_
```

```python
gridNN.cv_results_["mean_fit_time"].sum()/3600 # Time taken for grid search in hours
```

```python
# Creating lists of grid search results

optim = []
lr = []
crit = []
mean_test_score = []
std_test_score = []
for i in range(len(gridNN.cv_results_["mean_test_score"])):
    if str(gridNN.cv_results_["param_optimizer"][i]) == "<class 'torch.optim.sgd.SGD'>":
        optim.append("SGD")
    else:
        optim.append("Adam")
        
    lr.append(gridNN.cv_results_["param_lr"][i])
    
    if str(gridNN.cv_results_["param_criterion"][i]) == "<class 'torch.nn.modules.loss.CrossEntropyLoss'>":
        crit.append("CrossEntropyLoss")
    else:
        crit.append("NLLLoss")
        
    mean_test_score.append(gridNN.cv_results_["mean_test_score"][i])
    std_test_score.append(gridNN.cv_results_["std_test_score"][i])
```

```python
# Dataframe of grid search results
gs_result = pd.DataFrame(data={'optimizer': optim, 'learning rate': lr, 'loss': crit, 'mean score': mean_test_score, 'std score': std_test_score})
```

```python
gs_result = gs_result.sort_values("optimizer").reset_index().drop("index", axis = 1)
```

* lr: 0.0001

## Final model training and evaluation

```python
import torch.nn.functional as F

# Training model on optimal hyperparameters 

class FinalNet(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, dropout = 0.5):
        
        super(FinalNet, self).__init__()
        
        self.input = nn.Linear(input_dim, hidden_dim)
        self.hidden1 = nn.Linear(hidden_dim, int(hidden_dim/2))
        self.output = nn.Linear(int(hidden_dim/2), output_dim)
        
        self.dropout = nn.Dropout(dropout)
    
    def forward(self, X):
        
        input = self.input(X)
        input = F.relu(input)
        
        hidden1 = self.hidden1(input)
        hidden1 = F.relu(hidden1)
        
        output = self.output(hidden1)
        output = self.dropout(output)
        output = F.softmax(output, dim = -1)
        
        return output
```

```python
FNet = NeuralNetClassifier(
    FinalNet(inc_dim, hidden_dim, output_dim),
    max_epochs=300,
    callbacks=[EarlyStopping()],
    optimizer = torch.optim.Adam,
    criterion = nn.NLLLoss,
    lr = 0.00001
)
```
